# Лабораторная работа №1
## Необходимые и достаточные условия существования условного экстремума

### Задание 1.2

Найти экстремум функции:

$$f(x) = (x_1^3 - x_2)^2 + x_2^2 \to extr$$

при ограничениях:
$$g_1(x) = x_1^2 + x_2^2 - 1 \leq 0$$
$$g_2(x) = -x_1 \leq 0$$

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import Normalize
import warnings
warnings.filterwarnings('ignore')

print('Библиотеки загружены.')
```

## Шаг 1. Постановка задачи — символьные переменные и функции

```python
x1, x2 = sp.symbols('x1 x2', real=True)
lam1, lam2 = sp.symbols('lambda1 lambda2', real=True)

# Целевая функция
f = (x1**3 - x2)**2 + x2**2

# Ограничения (вид g(x) <= 0)
g1 = x1**2 + x2**2 - 1
g2 = -x1

print('Целевая функция:')
sp.pprint(f)
print('\nОграничение g1(x) =', g1, '<= 0')
print('Ограничение g2(x) =', g2, '<= 0')
```

## Шаг 2. Функция Лагранжа

Классическая функция Лагранжа (при $\lambda_0 = 1$):
$$L(x, \lambda) = f(x) + \lambda_1 g_1(x) + \lambda_2 g_2(x)$$

```python
L = f + lam1 * g1 + lam2 * g2

print('Функция Лагранжа L(x, λ):')
sp.pprint(sp.expand(L))

# Частные производные
dL_dx1 = sp.diff(L, x1)
dL_dx2 = sp.diff(L, x2)

print('\n∂L/∂x1 =')
sp.pprint(sp.expand(dL_dx1))

print('\n∂L/∂x2 =')
sp.pprint(sp.expand(dL_dx2))
```

## Шаг 3. Необходимые условия экстремума первого порядка (KKT)

**А) Условие стационарности:**
$$\frac{\partial L}{\partial x_i} = 0, \quad i = 1, 2$$

**Б) Условие допустимости:** $g_j(x^e) \leq 0$

**В) Условие неотрицательности (для минимума):** $\lambda_j \geq 0$

**Г) Условие дополняющей нежёсткости:** $\lambda_j g_j(x^e) = 0$

```python
print('='*60)
print('СЛУЧАЙ 1: оба ограничения пассивны — g1 < 0, g2 < 0')
print('  => λ1 = 0, λ2 = 0 (дополняющая нежёсткость)')
print('='*60)

# Подставляем λ1=0, λ2=0 -> условие стационарности обычной f
dL_dx1_00 = dL_dx1.subs([(lam1, 0), (lam2, 0)])
dL_dx2_00 = dL_dx2.subs([(lam1, 0), (lam2, 0)])

sol_case1 = sp.solve([dL_dx1_00, dL_dx2_00], [x1, x2])
print('\nСтационарные точки (внутри области):', sol_case1)

for sol in sol_case1:
    pt_x1, pt_x2 = sol
    v1 = g1.subs([(x1, pt_x1), (x2, pt_x2)])
    v2 = g2.subs([(x1, pt_x1), (x2, pt_x2)])
    fval = f.subs([(x1, pt_x1), (x2, pt_x2)])
    print(f'  Точка ({pt_x1}, {pt_x2}): g1={v1}, g2={v2}, f={fval}')
    feasible = (v1 <= 0) and (v2 <= 0)
    print(f'  Допустима: {feasible}')
```

```python
print('='*60)
print('СЛУЧАЙ 2: g1 активно, g2 пассивно')
print('  => g1(x)=0, λ2=0, λ1 >= 0')
print('='*60)

dL_dx1_c2 = dL_dx1.subs(lam2, 0)
dL_dx2_c2 = dL_dx2.subs(lam2, 0)

# Система: dL/dx1=0, dL/dx2=0, g1=0
sol_case2 = sp.solve([dL_dx1_c2, dL_dx2_c2, g1], [x1, x2, lam1])
print('\nРешения системы:', sol_case2)

valid_case2 = []
for sol in sol_case2:
    pt_x1, pt_x2, pt_lam1 = sol
    pt_x1 = sp.simplify(pt_x1)
    pt_x2 = sp.simplify(pt_x2)
    pt_lam1 = sp.simplify(pt_lam1)
    v2 = g2.subs([(x1, pt_x1), (x2, pt_x2)])
    fval = f.subs([(x1, pt_x1), (x2, pt_x2)])
    lam1_ok = (pt_lam1 >= 0)
    g2_ok = (v2 <= 0)
    print(f'  Точка ({pt_x1:.4f}, {pt_x2:.4f}): λ1={float(pt_lam1):.4f}, g2={float(v2):.4f}, f={float(fval):.4f}')
    print(f'  λ1>=0: {bool(lam1_ok)}, g2<=0: {bool(g2_ok)}')
    if bool(lam1_ok) and bool(g2_ok):
        valid_case2.append((pt_x1, pt_x2, pt_lam1, 0, fval))
        print('  ✓ KKT выполнены')
    else:
        print('  ✗ KKT не выполнены')
```

```python
print('='*60)
print('СЛУЧАЙ 3: g1 пассивно, g2 активно')
print('  => λ1=0, g2(x)=0 => x1=0, λ2 >= 0')
print('='*60)

# g2(x) = -x1 = 0 => x1 = 0
dL_dx1_c3 = dL_dx1.subs([(lam1, 0), (x1, 0)])
dL_dx2_c3 = dL_dx2.subs([(lam1, 0), (x1, 0)])

# Из dL/dx1=0 находим λ2, из dL/dx2=0 находим x2
sol_x2 = sp.solve(dL_dx2_c3, x2)
print('x2 из dL/dx2=0:', sol_x2)

for x2_val in sol_x2:
    lam2_val = sp.solve(dL_dx1_c3.subs(x2, x2_val), lam2)
    for l2 in lam2_val:
        v1 = g1.subs([(x1, 0), (x2, x2_val)])
        fval = f.subs([(x1, 0), (x2, x2_val)])
        print(f'  Точка (0, {x2_val}): λ2={l2}, g1={v1}, f={fval}')
        print(f'  λ2>=0: {bool(l2 >= 0)}, g1<=0: {bool(v1 <= 0)}')
```

```python
print('='*60)
print('СЛУЧАЙ 4: оба ограничения активны')
print('  => g1(x)=0, g2(x)=0 => x1=0, x1²+x2²=1 => x2=±1')
print('='*60)

# x1=0, g1=0 => x2=±1
for x2_val in [1, -1]:
    # Из системы стационарности находим λ1, λ2
    eq1 = dL_dx1.subs([(x1, 0), (x2, x2_val)])
    eq2 = dL_dx2.subs([(x1, 0), (x2, x2_val)])
    lam_sol = sp.solve([eq1, eq2], [lam1, lam2])
    fval = f.subs([(x1, 0), (x2, x2_val)])
    print(f'\n  Точка (0, {x2_val}): f={fval}')
    if lam_sol:
        l1 = lam_sol[lam1]
        l2 = lam_sol[lam2]
        print(f'  λ1={l1}, λ2={l2}')
        print(f'  λ1>=0: {bool(l1 >= 0)}, λ2>=0: {bool(l2 >= 0)}')
    else:
        print('  Нет решения для множителей')
```

## Шаг 4. Достаточные условия — второй дифференциал функции Лагранжа

Матрица Гессе функции Лагранжа:
$$H = \begin{pmatrix} \frac{\partial^2 L}{\partial x_1^2} & \frac{\partial^2 L}{\partial x_1 \partial x_2} \\ \frac{\partial^2 L}{\partial x_2 \partial x_1} & \frac{\partial^2 L}{\partial x_2^2} \end{pmatrix}$$

Второй дифференциал: $d^2L = dx^T H dx$

```python
# Матрица Гессе функции Лагранжа
H = sp.Matrix([
    [sp.diff(L, x1, x1), sp.diff(L, x1, x2)],
    [sp.diff(L, x2, x1), sp.diff(L, x2, x2)]
])

print('Матрица Гессе функции Лагранжа H(x, λ):')
sp.pprint(H)

# Градиенты ограничений
grad_g1 = sp.Matrix([sp.diff(g1, x1), sp.diff(g1, x2)])
grad_g2 = sp.Matrix([sp.diff(g2, x1), sp.diff(g2, x2)])

print('\n∇g1 =', grad_g1.T)
print('∇g2 =', grad_g2.T)
```

```python
def check_second_order(pt_x1, pt_x2, pt_lam1, pt_lam2, active_constraints, label):
    """Проверка достаточных условий второго порядка в точке."""
    print(f'\n--- {label}: ({float(pt_x1):.4f}, {float(pt_x2):.4f}) ---')
    
    # Гессе в точке
    H_pt = H.subs([(x1, pt_x1), (x2, pt_x2), (lam1, pt_lam1), (lam2, pt_lam2)])
    H_pt = sp.simplify(H_pt)
    print('H в точке:')
    sp.pprint(H_pt)
    
    # Условия dg=0 для активных ограничений
    dx1_s, dx2_s = sp.symbols('dx1 dx2')
    constraints_dx = []
    for g, grad_g in active_constraints:
        grad_pt = grad_g.subs([(x1, pt_x1), (x2, pt_x2)])
        eq = grad_pt[0]*dx1_s + grad_pt[1]*dx2_s
        constraints_dx.append(eq)
        print(f'dg = {eq} = 0')
    
    if len(constraints_dx) == 0:
        # Нет активных ограничений — проверяем знак H напрямую
        eigvals = H_pt.eigenvals()
        print('Собственные значения H:', eigvals)
    elif len(constraints_dx) == 1:
        # Одно активное ограничение: выражаем одну переменную через другую
        sol = sp.solve(constraints_dx[0], dx1_s)
        if sol:
            dx1_expr = sol[0]
            dx2_free = dx2_s
            d2L_expr = (sp.Matrix([dx1_expr, dx2_free]).T * H_pt * sp.Matrix([dx1_expr, dx2_free]))[0]
            d2L_simplified = sp.expand(d2L_expr)
            print(f'd²L при dx1={dx1_expr}:')
            sp.pprint(d2L_simplified)
            # Знак при dx2^2
            coeff = d2L_simplified.coeff(dx2_s**2)
            print(f'Коэффициент при dx2²: {coeff}')
            if coeff > 0:
                print('=> d²L > 0 => ЛОКАЛЬНЫЙ МИНИМУМ')
            elif coeff < 0:
                print('=> d²L < 0 => ЛОКАЛЬНЫЙ МАКСИМУМ')
            else:
                print('=> Знак неопределён, требуется дополнительный анализ')
        else:
            sol2 = sp.solve(constraints_dx[0], dx2_s)
            if sol2:
                dx2_expr = sol2[0]
                d2L_expr = (sp.Matrix([dx1_s, dx2_expr]).T * H_pt * sp.Matrix([dx1_s, dx2_expr]))[0]
                d2L_simplified = sp.expand(d2L_expr)
                print(f'd²L при dx2={dx2_expr}:')
                sp.pprint(d2L_simplified)
                coeff = d2L_simplified.coeff(dx1_s**2)
                print(f'Коэффициент при dx1²: {coeff}')
                if coeff > 0:
                    print('=> d²L > 0 => ЛОКАЛЬНЫЙ МИНИМУМ')
                elif coeff < 0:
                    print('=> d²L < 0 => ЛОКАЛЬНЫЙ МАКСИМУМ')
                else:
                    print('=> Знак неопределён')
    elif len(constraints_dx) == 2:
        # Два активных ограничения — пространство dx одномерно или нулевое
        sol = sp.solve(constraints_dx, [dx1_s, dx2_s])
        print('Допустимые dx:', sol)
        if sol == {dx1_s: 0, dx2_s: 0} or sol == []:
            print('=> Нет ненулевых допустимых dx, достаточные условия выполняются тривиально')
        else:
            print('=> Требуется дополнительный анализ')
```

```python
print('ПРОВЕРКА ДОСТАТОЧНЫХ УСЛОВИЙ 2-го ПОРЯДКА')
print('='*60)

# Случай 1: (0, 0), λ1=0, λ2=0 — нет активных ограничений
check_second_order(0, 0, 0, 0, [], 'Случай 1: точка (0,0), λ1=0, λ2=0')
f_val_00 = float(f.subs([(x1, 0), (x2, 0)]))
print(f'f(0,0) = {f_val_00}')
```

```python
# Случай 2: активно g1, пассивно g2 - используем найденные точки
print('СЛУЧАЙ 2: активно g1 = 0')
print('Проверяем точки с границы окружности x1²+x2²=1, x1>=0')

# Подробный анализ точек с границы g1=0
if valid_case2:
    for pt in valid_case2:
        pt_x1_v, pt_x2_v, pt_lam1_v, pt_lam2_v, fval = pt
        check_second_order(
            pt_x1_v, pt_x2_v, pt_lam1_v, pt_lam2_v,
            [(g1, grad_g1)],
            f'Случай 2: точка ({float(pt_x1_v):.4f}, {float(pt_x2_v):.4f})'
        )
        print(f'f = {float(fval):.6f}')
else:
    print('Нет допустимых KKT-точек в случае 2')
```

```python
# Случай 4: обе активны — точки (0, 1) и (0, -1)
print('СЛУЧАЙ 4: активны g1 и g2 — точки (0,1) и (0,-1)')
for x2_val in [1, -1]:
    eq1 = dL_dx1.subs([(x1, 0), (x2, x2_val)])
    eq2 = dL_dx2.subs([(x1, 0), (x2, x2_val)])
    lam_sol = sp.solve([eq1, eq2], [lam1, lam2])
    if lam_sol and lam1 in lam_sol and lam2 in lam_sol:
        l1_v = lam_sol[lam1]
        l2_v = lam_sol[lam2]
        check_second_order(
            0, x2_val, l1_v, l2_v,
            [(g1, grad_g1), (g2, grad_g2)],
            f'Случай 4: точка (0, {x2_val})'
        )
        print(f'f(0, {x2_val}) = {float(f.subs([(x1,0),(x2,x2_val)]))}')
```

## Шаг 5. Сводная таблица результатов

```python
print('='*70)
print('СВОДНАЯ ТАБЛИЦА СТАЦИОНАРНЫХ ТОЧЕК')
print('='*70)
print(f'{"Точка":>20} | {"f(x)":>12} | {"Тип":>20}')
print('-'*70)

# Все потенциальные KKT точки
candidates = [
    ((0, 0),   float(f.subs([(x1,0),(x2,0)])),   'Глобальный минимум'),
    ((0, 1),   float(f.subs([(x1,0),(x2,1)])),   'Граничная точка'),
    ((0, -1),  float(f.subs([(x1,0),(x2,-1)])),  'Граничная точка'),
]

# Добавим точки из случая 2
if valid_case2:
    for pt in valid_case2:
        px, py = float(pt[0]), float(pt[1])
        fv = float(pt[4])
        candidates.append(((px, py), fv, 'KKT-точка (граница g1)'))

for pt, fv, tp in candidates:
    print(f'({pt[0]:>7.4f}, {pt[1]:>7.4f}) | {fv:>12.6f} | {tp}')

fvals = [c[1] for c in candidates]
print('-'*70)
print(f'Минимальное значение f = {min(fvals):.6f}')
print(f'Максимальное значение f = {max(fvals):.6f}')
```

## Шаг 6. Визуализация

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# --- Сетка ---
x1_range = np.linspace(-0.1, 1.1, 400)
x2_range = np.linspace(-1.2, 1.2, 400)
X1, X2 = np.meshgrid(x1_range, x2_range)
F = (X1**3 - X2)**2 + X2**2
G1 = X1**2 + X2**2 - 1
G2 = -X1

# Допустимая область: G1 <= 0 AND G2 <= 0
feasible = (G1 <= 0) & (G2 <= 0)

# --- График 1: Линии уровня + допустимая область ---
ax = axes[0]
F_masked = np.where(feasible, F, np.nan)

cp = ax.contourf(X1, X2, F, levels=40, cmap='viridis', alpha=0.6)
fig.colorbar(cp, ax=ax, label='f(x)')
ax.contour(X1, X2, F, levels=20, colors='white', alpha=0.3, linewidths=0.5)

# Допустимая область — полупрозрачная заливка
ax.contourf(X1, X2, feasible.astype(float), levels=[0.5, 1.5],
            colors=['cyan'], alpha=0.25)
ax.contour(X1, X2, G1, levels=[0], colors='red', linewidths=2, linestyles='-')
ax.contour(X1, X2, G2, levels=[0], colors='orange', linewidths=2, linestyles='--')

# Отмечаем стационарные точки
points_to_plot = [
    (0, 0,   'Минимум (0,0)\nf=0',    'lime',   200, '*'),
    (0, 1,   '(0,1)\nf=1',            'yellow', 100, 'o'),
    (0, -1,  '(0,-1)\nf=1',           'yellow', 100, 'o'),
]
if valid_case2:
    for pt in valid_case2:
        px, py, _, _, fv = [float(v) for v in pt]
        points_to_plot.append((px, py, f'({px:.3f},{py:.3f})\nf={fv:.3f}', 'magenta', 120, 's'))

for px, py, lbl, clr, sz, mk in points_to_plot:
    ax.scatter(px, py, color=clr, s=sz, marker=mk, zorder=5)
    ax.annotate(lbl, (px, py), textcoords='offset points', xytext=(8, 6),
                fontsize=8, color='white',
                bbox=dict(boxstyle='round,pad=0.2', fc='black', alpha=0.5))

ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-1.2, 1.2)
ax.set_xlabel('x₁')
ax.set_ylabel('x₂')
ax.set_title('Линии уровня f(x) и допустимая область')
ax.set_aspect('equal')
patches = [
    mpatches.Patch(color='cyan', alpha=0.5, label='Допустимая область'),
    mpatches.Patch(color='red', label='g₁(x)=0 (граница)'),
    mpatches.Patch(color='orange', label='g₂(x)=0 (граница)'),
]
ax.legend(handles=patches, fontsize=8, loc='upper right')
ax.grid(True, alpha=0.3)

# --- График 2: 3D поверхность ---
ax3d = fig.add_subplot(1, 2, 2, projection='3d')
axes[1].remove()

x1_3d = np.linspace(0, 1, 80)
x2_3d = np.linspace(-1, 1, 80)
X1_3d, X2_3d = np.meshgrid(x1_3d, x2_3d)
F_3d = (X1_3d**3 - X2_3d)**2 + X2_3d**2
G1_3d = X1_3d**2 + X2_3d**2 - 1
F_3d_masked = np.where(G1_3d <= 0, F_3d, np.nan)

surf = ax3d.plot_surface(X1_3d, X2_3d, F_3d_masked, cmap='plasma',
                          alpha=0.8, linewidth=0, antialiased=True)
ax3d.scatter(0, 0, 0, color='lime', s=100, zorder=10, label='Минимум (0,0,0)')
ax3d.set_xlabel('x₁')
ax3d.set_ylabel('x₂')
ax3d.set_zlabel('f(x)')
ax3d.set_title('Поверхность f(x) на допустимом множестве')
ax3d.legend(fontsize=8)

plt.suptitle('Задание 1.2: f(x) = (x₁³ - x₂)² + x₂² → extr', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('task1_2_plot.png', dpi=150, bbox_inches='tight')
plt.show()
print('График сохранён в task1_2_plot.png')
```

## Вывод

**Глобальный минимум:** $f(0, 0) = 0$ в точке $x^* = (0, 0)$.

- Точка $(0,0)$ находится **внутри** допустимого множества: $g_1(0,0) = -1 < 0$, $g_2(0,0) = 0 \leq 0$.
- Все ограничения пассивны (кроме граничного $g_2$), $\lambda_1 = \lambda_2 = 0$.
- $f(0,0) = 0$ — глобальный минимум, поскольку $f(x) \geq 0$ для всех $x$.
- Матрица Гессе $f$ в точке $(0,0)$ положительно полуопределена, что подтверждает минимум.
